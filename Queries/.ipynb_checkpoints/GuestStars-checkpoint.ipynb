```python
# introduce dependencies
import pandas as pd
import datetime as dt
```

```python
# read in csv and convert to dataframe
guest_txt = "../Data/guest-stars.txt"
guest_df = pd.read_csv(guest_txt, sep = 'delimiter', encoding = "utf-8", header = None)
# guest_df = pd.read_table(guest_txt, encoding = "utf-8")
guest_df.head()
```

```python
show_df = guest_df[0].str.split(',', expand=True).add_prefix("Episode_")
show_df.rename(columns=({"Episode_0": "Actor"}), inplace=True)
show_df
```

```python
guest_transpose = show_df.transpose()
guest_transpose.reset_index(inplace = True)
guest_transpose.rename(columns = ({"index": "episode_number"}), inplace = True)
guest_transpose.drop(0, axis = 0, inplace = True)
show_list = guest_transpose["episode_number"].unique()
show_list
```

```python
guest_show_df = pd.melt(show_df, id_vars = "Actor", value_vars= show_list)
guest_show_df.dropna(inplace=True)
```

```python
guest_show_df.drop(labels = "variable", axis=1, inplace = True)
guest_show_df.rename(columns = ({"value": "Title"}), inplace=True)
guest_show_df.sort_values(by = ["Actor"],  inplace=True)
guest_show_df
```
